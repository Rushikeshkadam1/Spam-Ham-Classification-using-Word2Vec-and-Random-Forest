```python
import pandas as pd
spam_or_ham = "/content/spamhamdata.csv"
messages = pd.read_csv(spam_or_ham, sep='\t', names=["label", "message"])
```

```python
from nltk.stem import WordNetLemmatizer
import re
import nltk
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range (0, len(messages)):
    review = re.sub('[^a-zA-Z]',' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word, pos ='v') for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)), corpus, messages['message']) if i <1]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
'''  simple_preprocess() -
# Convert a document into a list of lowercase tokens, ignoring tokens that are too short or too long.
'''
```

```python
words = []
for sentence in corpus:
  sentence_token = sent_tokenize(sentence)
  for sentence in sentence_token:
    words.append(simple_preprocess(sentence))
```

```python
# Word2Vec is giving 100D vector for every each and every word in sentence
# but we want a 100D vector for a single sentence
# hence We are using AvgWord2Vec
```

```python
def avgword2vec(doc):
    # Filter words that are in the model's vocabulary
    word_vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]

    # If there are no valid word vectors, return a zero vector of the correct dimension
    if len(word_vectors) == 0:
        return np.zeros(model.vector_size)

    # Compute the mean of the word vectors along axis 0 (to get a 100D vector)
    return np.mean(word_vectors, axis=0)
```
